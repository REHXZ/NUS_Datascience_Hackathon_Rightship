```python
import matplotlib.pyplot as plt

# Increase the figure size
plt.figure(figsize=(12, 6))  # Adjust width and height as needed

# Your current plot
ax = (
    ais_dataset["speed"]
    .value_counts()
    .plot(kind="bar", title="Distribution of speed", width=0.8)  # Adjust bar width
)

# Add values above the bars
for p in ax.patches:
    ax.annotate(
        f"{int(p.get_height())}",  # Text to display (value of the bar)
        (p.get_x() + p.get_width() / 2, p.get_height()),  # Position of the text
        ha="center",  # Horizontal alignment
        va="bottom",  # Vertical alignment
        xytext=(0, 5),  # Offset from the bar
        textcoords="offset points",  # Coordinate system for the offset
    )

# Add more space between the bars
plt.subplots_adjust(wspace=0.5)  # Adjust horizontal space between bars

# Show the plot
plt.show()
```
